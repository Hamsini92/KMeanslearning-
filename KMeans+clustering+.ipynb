```python
df['tokenized_sents'] = df.apply(lambda row: nltk.word_tokenize(row['name']), axis=1)
```

```python
#perform the preprocessing to remove the stop words in English 
stop = stopwords.words('english')
df['Transaction'] = df['tokenized_sents'].apply(lambda x: [item for item in x if item not in stop])
```

```python
documents = df.values.T.tolist()
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

documents = ["Bill, Payment, CITI, CARDS, CBOL, 10021.",
             "ACH, Electronic, Debit, -, DISCOVER, E-PAYMEN.",
             "STARS, STRIPES, EXPRESS, C].",
             "HABIT-MIRA, MESA, #, 135, MIRA.",
             "IN, N, OUT, BURGER, 095, KETTLEMAN, CITCA",
             "Online, Banking, Transfer, Conf, #, 37m3vd6zp.",
             "Online, Banking, payment, CSL, 0386, Confirmati",
             "ACH, Electronic, Debit, -, CA102-, Avalon, Mo."]

#documents = df['Transaction']

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

true_k = 3
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)

print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print

print("\n")
print("Prediction")

Y = vectorizer.transform(["ACH, Electronic, Debit, -, CA102-, Avalon, Mo."])
prediction = model.predict(Y)
print(prediction)

Y = vectorizer.transform(["Online, Banking, Transfer, Conf."])
prediction = model.predict(Y)
print(prediction)
```
